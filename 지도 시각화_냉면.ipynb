```python
import googlemaps
gmaps = googlemaps.Client(key="받은 API KEY") #본인 APi넣어서 사용하시면 됩니다!
```

```python
#위도와 경도 값 추가
lat = []
lng = []

for name in restaurant_df['주소']:
    tmpMap = gmaps.geocode(name)
    tmpLoc = tmpMap[0].get('geometry') # 배열 형태( [~] )로 들어오기 때문에 [0]으로 호출
    lat.append(tmpLoc['location']['lat']) # dict(tmpLoc)의 데이터는 dict['key값'] 로 value 호출
    lng.append(tmpLoc['location']['lng'])
    
restaurant_df['lat'] = lat
restaurant_df['lng'] = lng

restaurant_df.head()
```

```python
#망고평점(값이 비슷해 점수로 환산) 추가

def reRange(x, oldMin, oldMax, newMin, newMax):
    return (x - oldMin)*(newMax - newMin) / (oldMax - oldMin) + newMin 

restaurant_df['망고환산'] = reRange(restaurant_df['망고평점'], min(restaurant_df['망고평점']), max(restaurant_df['망고평점']), 1, 100)
restaurant_df.head()
```

```python
#다이닝 코드별점(값이 비슷해 점수로 환산) 추가 ->다이닝평점(100점기준)도 있으나 망고 평점(5점기준)과의 비교를 위해 다이닝 코드 별점만 구함

def reRange(x, oldMin, oldMax, newMin, newMax):
    return (x - oldMin)*(newMax - newMin) / (oldMax - oldMin) + newMin 

restaurant_df['다코환산'] = reRange(restaurant_df['다이닝코드별점'], min(restaurant_df['다이닝코드별점']), max(restaurant_df['다이닝코드별점']), 1, 100)
restaurant_df.head()
```

```python
#서울시 자료만 모으기

restaurant_df_seoul = restaurant_df[ restaurant_df['주소'].str.startswith('서울') ]
```

```python
#구별 값 추출

restaurant_df_seoul['구별'] = restaurant_df_seoul['주소'].apply(lambda x : x.split(' ')[1])
restaurant_df_seoul.head(3)
```

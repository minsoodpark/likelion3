```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt 
```

```python
lat = []
lng = []

for name in restaurant_seoul['주소']:
    # gmaps.reverse_geocode((longitude 값, latitude 값), language="ko") == 경도 & 위도 값으로 주소값 가져오기
    # gmaps.geocode('한글 주소', language="ko") 로 위도/경도, 우편번호까지 알 수 있음
    tmpMap = gmaps.geocode(name) # ex) 서울강남경찰서
    tmpLoc = tmpMap[0].get('geometry') # 배열 형태( [~] )로 들어오기 때문에 [0]으로 호출
    lat.append(tmpLoc['location']['lat']) # dict(tmpLoc)의 데이터는 dict['key값'] 로 value 호출
    lng.append(tmpLoc['location']['lng'])
    
restaurant_seoul['lat'] = lat
restaurant_seoul['lng'] = lng

restaurant_seoul.head(3)
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11)

for n in restaurant_seoul.index: 
    
    folium.CircleMarker ([restaurant_seoul['lat'][n], restaurant_seoul['lng'][n]], # n 자리는 인덱스 번호  # 행,열 주의하기
                         radius=restaurant_seoul['다이닝코드평점'][n]*0.5, # circle 의 크기를 결정
                         color='#3186cc', fill=True, fill_color='#3186cc').add_to(map) #지도에 얹어줌.

map

# df['lat'][n] [행][열]
# df.loc['lat'][n] [열][행]
```

```python
def get_mango_info(restaurant):#ìƒí˜¸ëª…ì„ inputìœ¼ë¡œ ê´€ë ¨ ì •ë³´ ë°›ì•„ì˜¤ëŠ” function

    mango_info = {} #ë¹ˆ dict ìƒì„±

    url = 'https://www.mangoplate.com/search/'+restaurant #ìƒí˜¸ëª…ìœ¼ë¡œ ë§ê³ í”Œë ˆì´íŠ¸ ê²€ìƒ‰ url

    session = requests.Session()
    headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5) AppleWebKit 537.36 (KHTML, like Gecko) Chrome",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8"
    }
    html = session.get(url, headers=headers).content 
    source = BeautifulSoup(html, 'html.parser')
    
    try:
        rest_code = source.find('div',{'class':'list-restaurant-item'}).find('a')['href']#ê²€ìƒ‰ê²°ê³¼ì—ì„œ ì²«ë²ˆì§¸ ê²°ê³¼ì˜ ê³ ìœ  ì½”ë“œ ë”°ì˜¤ê¸° 
        newurl = 'https://www.mangoplate.com'+rest_code #ê³ ìœ  ì½”ë“œ í™œìš©í•œ url ì£¼ì†Œ ìƒì„±
        newhtml = session.get(newurl, headers=headers).content 
        newsource = BeautifulSoup(newhtml, 'html.parser')

        rate = float(re.sub('\s','',newsource.find('strong',{'class':'rate-point'}).get_text()))#í‰ì ì´ ì†í•œ íƒœê·¸ì—ì„œ í…ìŠ¤íŠ¸ ê°€ì ¸ì™€ floatë¡œ ë°”ê¾¸ê¸° (ë‚˜ì¤‘ì— ê³„ì‚° ìš©ì´)

        rest_table = newsource.find('table',{'class':'info'})#ê´€ë ¨ ì •ë³´ê°€ ìˆëŠ” í…Œì´ë¸” ê²½ë¡œ

        address = rest_table.find_all('td')[0].find_all('span')[1].get_text()#ì£¼ì†Œ
        phone = rest_table.find_all('td')[1].get_text()#ì „í™”ë²ˆí˜¸
        kind = re.sub('\s','',rest_table.find_all('td')[2].get_text())#ìŒì‹ ì¢…ë¥˜(ì •ê·œí‘œí˜„ì‹ í™œìš©í•˜ì—¬ ê¹”ë”í•˜ê²Œ ì •ë¦¬)
        price = rest_table.find_all('td')[3].get_text()#ê°€ê²©ëŒ€

        #ê°ê°ì˜ ì •ë³´ë¥¼ ë§Œë“  dictì˜ keyê°’ì˜ valueë¡œ ì¶”ê°€
        mango_info['name'] = restaurant
        mango_info['rate'] = rate
        mango_info['address'] = address
        mango_info['phone'] = phone
        mango_info['kind'] = kind
        mango_info['price'] = price
        
    except:
        mango_info['name'] = restaurant
        mango_info['rate'] = 'N/A'
        mango_info['address'] = 'N/A'
        mango_info['phone'] = 'N/A'
        mango_info['kind'] = 'N/A'
        mango_info['price'] = 'N/A'
    
    return mango_info #í•˜ë‚˜ì˜ dictë¥¼ ìµœì¢… ê°’ìœ¼ë¡œ return

```

```python
def get_naver_info(restaurant): #ë„¤ì´ë²„ í‰ì  ë° ë¦¬ë·° ê°€ì ¸ì˜¤ê¸°
    
    naver_info = {}
    
    #ì •ê·œí‘œí˜„ì‹ í™œìš©í•´ ìƒí˜¸ëª…ì— ìˆëŠ” ê´„í˜¸ ë° íŠ¹ìˆ˜ë¬¸ì ì œê±°
    regex = "\(.*\)|\s-\s.*" 
    query = re.sub(regex, '', restaurant)
    
    url = "https://search.naver.com/search.naver?where=nexearch&sm=tab_jum&query=" + query
    web = requests.get(url).content
    source = BeautifulSoup(web, 'html.parser')
    
    try:
        reviewcount = int(source.find_all('span', {'class': 'count'})[1].get_text())
        rate = float(source.find_all('div', {'class': 'star_area'})[0].find('em').get_text())
    except:
        reviewcount = 'N/A'
        rate = 'N/A'
    
    naver_info['reviewcount'] = reviewcount
    naver_info['rate'] = rate
    
    return naver_info


    
```

```python
def get_dining_info(restaurant): #ë‹¤ì´ë‹ì½”ë“œ í‰ì /ë³„ì /ë¦¬ë·° ê°€ì ¸ì˜¤ê¸°
    
    dining_info = {}
    review_list = []
    
    #ì •ê·œí‘œí˜„ì‹ í™œìš©í•´ ìƒí˜¸ëª…ì— ìˆëŠ” ê´„í˜¸ ë° íŠ¹ìˆ˜ë¬¸ì ì œê±°
    regex = "\(.*\)|\s-\s.*"
    query = re.sub(regex, '', restaurant)
    
    url = 'https://www.diningcode.com/isearch.php?query=' + query
    web = requests.get(url).content
    source = BeautifulSoup(web, 'html.parser')
    
    #ë‹¤ì´ë‹ì½”ë“œì— í•´ë‹¹ ì‹ë‹¹ì´ ì—†ëŠ” ê²½ìš°ë¥¼ ì˜ˆìƒí•˜ì—¬ try/exceptë¡œ error ë°œìƒ ì‹œ 'N/A'ë¡œ ë°ì´í„° ëŒ€ì²´
    try:
        ul_tag = source.find('div', {'id':'div_rn'})
        li_tag = ul_tag.find('li')
        a_tag = li_tag.find('a')
        target_url = 'https://www.diningcode.com' + a_tag.attrs['href']
        target_url

        web_review = requests.get(target_url).content
        source_review = BeautifulSoup(web_review, 'html.parser')

        #í‰ê·  í‰ì  êº¼ë‚´ê¸°
        p_tag = source_review.find('p', {'class':'grade'})
        # p_tag
        review_point = int(p_tag.find('strong').get_text().replace('ì ',''))

        #í‰ê·  ë³„ì  êº¼ë‚´ê¸°
        star_point = source_review.find('p', {'id':'lbl_star_point'}).get_text()
        star_point = star_point.replace("\n", "")
        star_point = float(star_point.replace('ì ',''))# êµì²´í•˜ë‹¤

        reviews = source_review.find_all('p',{'class':'review_contents btxt'})

        for review in reviews:
            review_list.append(re.sub('\n','',review.get_text()))

        dining_info['review_point'] = review_point
        dining_info['star_point'] = star_point
        dining_info['review'] = review_list
                               
    except:
        dining_info['review_point'] = 'N/A'
        dining_info['star_point'] = 'N/A'
        dining_info['review'] = 'N/A'
        
    return dining_info
```

```python
#ìŒì‹ì„ í‚¤ì›Œë“œë¡œ í•´ë‹¹ ìŒì‹ ë§›ì§‘ë¦¬ìŠ¤íŠ¸ë¥¼ ë½‘ì•„ì˜¤ëŠ” function
def get_restaurant_data(menu):
    
    #ë§Œë“¤ì–´ë†“ì€ get_rest_list functionìœ¼ë¡œ ìƒí˜¸ëª… ë¦¬ìŠ¤íŠ¸ ìƒì„±
    restaurant_list = get_rest_list(menu)
    print(len(restaurant_list),'ê°œì˜ ì‹ë‹¹ ê²€ìƒ‰ ì™„ë£Œ.')
    
    #DataFrameì„ ìœ„í•œ ë¹ˆ ë¦¬ìŠ¤íŠ¸ ìƒì„±

    mango_name = []
    mango_rate = []
    mango_address = []
    mango_phone = []
    mango_kind = []
    mango_price = []
    naver_reviewcount = []
    naver_rate = []
    dining_review_point = []
    dining_star_point = []
    dining_review = []
    count = 1

    #ìƒí˜¸ëª… ë¦¬ìŠ¤íŠ¸ë¥¼ í™œìš©í•˜ì—¬ ê°ê°ì˜ ì í¬ì˜ ì •ë³´ë¥¼ ë§ê³ í”Œë ˆì´íŠ¸, ë„¤ì´ë²„, ë‹¤ì´ë‹ì½”ë“œë¡œ ë¶€í„° ë°›ì•„ì˜¤ëŠ” for loop
    for restaurant in restaurant_list:

        mango_info = get_mango_info(restaurant)
        naver_info = get_naver_info(restaurant)
        dining_info = get_dining_info(restaurant)
        #ìœ„ì—ì„œ ì •ì˜í•œ ë§ê³ í”Œë ˆì´íŠ¸ ìŒì‹ì  ì •ë³´ë¥¼ í•˜ë‚˜ì˜ dictë¡œ ê°€ì ¸ì˜´

        #ê°ê°ì˜ ë¦¬ìŠ¤íŠ¸ì— ê³ ìœ  ì •ë³´ ì¶”ê°€
        mango_name.append(mango_info['name'])
        mango_rate.append(mango_info['rate'])
        mango_address.append(mango_info['address'])
        mango_phone.append(mango_info['phone'])
        mango_kind.append(mango_info['kind'])
        mango_price.append(mango_info['price'])
        naver_reviewcount.append(naver_info['reviewcount'])
        naver_rate.append(naver_info['rate'])
        dining_review_point.append(dining_info['review_point'])
        dining_star_point.append(dining_info['star_point'])
        dining_review.append(dining_info['review'])
        
        print(count,'ë²ˆì§¸ ì‹ë‹¹ ì •ë³´ ìˆ˜ì§‘ì¤‘...')
        count += 1
    
    #ìƒì„±ëœ ë¦¬ìŠ¤íŠ¸ë¡œ pandas dataframe ìƒì„±
    restaurant_df = pd.DataFrame({'ìŒì‹ì ':mango_name,
                         'ë§ê³ í‰ì ': mango_rate,
                         'ë„¤ì´ë²„í‰ì ': naver_rate,
                         'ë„¤ì´ë²„ë¦¬ë·°ìˆ˜': naver_reviewcount,
                         'ë‹¤ì´ë‹ì½”ë“œí‰ì ': dining_review_point,
                         'ë‹¤ì´ë‹ì½”ë“œë³„ì ': dining_star_point,
                         'ë‹¤ì´ë‹ì½”ë“œë¦¬ë·°': dining_review,     
                         'ì£¼ì†Œ': mango_address,
                         'ì „í™”': mango_phone,
                         'ì¢…ë¥˜': mango_kind,
                         'ê°€ê²©': mango_price})
    
    #ì™„ì„±ëœ DataFrameìœ¼ë¡œ excelíŒŒì¼ ìƒì„±
    restaurant_df.to_excel('{}ë§›ì§‘ë¦¬ìŠ¤íŠ¸_{}.xlsx'.format(menu,datetime.now().strftime('%y%m%d_%H%M')), index=False, encoding='utf-8')
    print(menu,'ë§›ì§‘ë¦¬ìŠ¤íŠ¸ê°€ ì™„ì„±ë˜ì—ˆìŠµë‹ˆë‹¤.')
    print(restaurant_df.head())
    
    #ì™„ì„±ëœ Dataframeì„ return
    return restaurant_df
    
```

```python
## ì‘ì—…ì¤‘

def food_word_cloud(food_name):
    
    df = pd.read_excel('{}.xlsx'.format(food_name), encoding='utf-8')
    
    food_dining_review = df['ë‹¤ì´ë‹ì½”ë“œë¦¬ë·°'].dropna()
    food_dining_review = list(food_dining_review)
    food_dining_review = ' '.join(food_dining_review)
    
    food_dining_review1 = food_dining_review[:6000]
    food_dining_review2 = food_dining_review[6000:12000]
    food_dining_review3 = food_dining_review[12000:18000]
    food_dining_review4 = food_dining_review[18000:]
    
    twitter = Twitter()
    raw_pos_tagged1 = twitter.pos(food_dining_review1, norm=True, stem=True) # POS Tagging
    raw_pos_tagged2 = twitter.pos(food_dining_review2, norm=True, stem=True) # POS Tagging
    raw_pos_tagged3 = twitter.pos(food_dining_review3, norm=True, stem=True) # POS Tagging
    raw_pos_tagged4 = twitter.pos(food_dining_review4, norm=True, stem=True) # POS Tagging
    
    raw_pos_tagged = []

    for raw_reviews in [food_dining_review1, food_dining_review2, food_dining_review3, food_dining_review4]:
        raw_pos_tagged += twitter.pos(raw_reviews, norm=True, stem=True)
        
    del_list = ['í•˜ë‹¤', 'ìˆë‹¤', 'ë˜ë‹¤', 'ì´ë‹¤', 'ë¼ë‹¤', 'ì•Šë‹¤', 'ê·¸ë ‡ë‹¤', 'ì•„ë‹ˆë‹¤', 'ì´ë ‡ë‹¤', 'ê·¸ë ‡ë‹¤', 'ì–´ë–»ë‹¤'] 

    word_cleaned = []
    for word in raw_pos_tagged: #  ('ì„œìš¸', 'Noun'),
        if not word[1] in ["Josa", "Eomi", "Punctuation", "Foreign"]: # Foreign == â€, â€œ ì™€ ê°™ì´ ì œì™¸ë˜ì–´ì•¼í•  í•­ëª©ë“¤
            if (len(word[0]) != 1) & (word[0] not in del_list): # í•œ ê¸€ìë¡œ ì´ë¤„ì§„ ë‹¨ì–´ë“¤ì„ ì œì™¸ & ì›ì¹˜ ì•ŠëŠ” ë‹¨ì–´ë“¤ì„ ì œì™¸
                word_cleaned.append(word[0])

    word_dic = {}

    for word in word_cleaned:
        if word not in word_dic:
            word_dic[word] = 1 # changed from "0" to "1"
        else:
            word_dic[word] += 1
    
    return word_dic
```
